```python
# Set a style for the plots
sns.set(style="whitegrid")
```

```python
# Histogram :-
# Question 1: Age Distribution :-
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax = sns.histplot(df['Age'], bins=20, kde=True, color="Orange")
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")
plt.show()
```

```python
# Boxplot
# Question 2: Gender-wise Data Usage
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.boxplot(data=df, x='Gender', y='Data Usage (GB/month)', palette='pastel')
ax.set_title('Gender-wise Data Usage')
ax.set_xlabel('Gender')
ax.set_ylabel('Data Usage (GB/month)')
plt.show()
```

```python
# countplot
# Question 3: Phone Brand Popularity
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.countplot(x='Phone Brand', data=df, palette='viridis', order=df['Phone Brand'].value_counts().index)
ax.set_title('Phone Brand Popularity')
ax.set_xlabel('Phone Brand')
ax.set_ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot.
# Question 4: Monthly Recharge Cost by Age Group
```

```python
age_bins = [0, 18, 30, 45, 60, 100]
labels = ['<18', '18-30', '30-45', '45-60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=labels)
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.barplot(data=df, x='Age Group', y='Monthly Recharge Cost (INR)', errorbar=None,
palette=sns.color_palette("coolwarm", len(labels)))
ax.set_title('Monthly Recharge Cost by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Monthly Recharge Cost (INR)')
plt.show()
```

```python
# Heatmap 
# Question 5: Heatmap of Correlations.
```

```python
fig, ax = plt.subplots(figsize=(9,5 ))
numeric_columns = df.select_dtypes(include=[np.number])
correlation = numeric_columns.corr()
ax = sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
ax.set_title('Heatmap of Feature Correlations')
plt.show()
```

```python
# Pie chart
# Question 6: Primary Use Distribution
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
primary_use_counts = df['Primary Use'].value_counts()
ax.pie(primary_use_counts, labels=primary_use_counts.index, autopct='%1.1f%%',
startangle=140, colors=sns.color_palette('Set2', len(primary_use_counts)))
ax.set_title('Primary Use Distribution (Pie Chart)')
plt.show()
```

```python
# Scatter plot
# Question 7: Social Media Time vs. Streaming Time
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.scatterplot(data=df, x='Social Media Time (hrs/day)', 
y='Streaming Time (hrs/day)', hue='Gender', palette='muted')
ax.set_title('Social Media Time vs. Streaming Time')
ax.set_xlabel('Social Media Time (hrs/day)')
ax.set_ylabel('Streaming Time (hrs/day)')
plt.show()
```

```python
# Box plot.
# Question 8: Calls Duration by Gender and Location
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.boxplot(data=df, x='Location',
y='Calls Duration (mins/day)', hue='Gender', palette='cool')
ax.set_title('Calls Duration by Gender and Location')
ax.set_xlabel('Location')
ax.set_ylabel('Calls Duration (mins/day)')
plt.xticks(rotation=45)
plt.show()
```

```python
# density plot
# Question 9: Gaming Time Distribution by OS
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.kdeplot(data=df, x='Gaming Time (hrs/day)',
hue='OS', fill=True, common_norm=False, palette='crest')
ax.set_title('Gaming Time Distribution by OS')
ax.set_xlabel('Gaming Time (hrs/day)')
ax.set_ylabel('Density')
plt.show()
```

```python
# scatter plot
# Question 10: E-commerce Spend vs. Data Usage
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax = sns.regplot(data=df, x='Data Usage (GB/month)', y='E-commerce Spend (INR/month)',
scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
ax.set_title('E-commerce Spend vs. Data Usage')
ax.set_xlabel('Data Usage (GB/month)')
ax.set_ylabel('E-commerce Spend (INR/month)')
plt.show()
```
